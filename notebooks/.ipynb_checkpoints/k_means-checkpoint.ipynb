```python
import numpy as np
```

```python
X = np.array([[1, 2], [0, 1], [1, 0], [4, 2], [4, 4], [3, 3]])
```

```python
kmeans = KMeans(n_clusters=2).fit(X)
kmeans.labels_
# array([0, 0, 0, 1, 1, 1], dtype=int32)
```

```python
kmeans.predict([4, 5])
# array([0, 1], dtype=int32)
```

```python
kmeans.cluster_centers_
# array([[1., 2.],[4., 2.]])
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data[:, 2:]
y = iris.target
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(X[y==0][:, 0], X[y==0][:, 1], color='r', label='0')
plt.scatter(X[y==1][:, 0], X[y==1][:, 1], color='y', label='1')
plt.scatter(X[y==2][:, 0], X[y==2][:, 1], color='g', label='2')
plt.legend();
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
```

```python
print(kmeans.labels_)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[kmeans.labels_==0][:, 0], X[kmeans.labels_==0][:, 1], color='r', label='0')
plt.scatter(X[kmeans.labels_==1][:, 0], X[kmeans.labels_==1][:, 1], color='y', label='1')
plt.scatter(X[kmeans.labels_==2][:, 0], X[kmeans.labels_==2][:, 1], color='g', label='2')
plt.legend();
```
